    ```python
    if —É–º–æ–≤–∞:

    elif —ñ–Ω—à–∞_—É–º–æ–≤–∞:
        # –¥—ñ—ó, —è–∫—â–æ –≤–∏–∫–æ–Ω–∞–Ω–∞ —ñ–Ω—à–∞ —É–º–æ–≤–∞
    else:

    ```

    ```python
    for i in range(5):
    ```

   - `break` ‚Äî –ø–µ—Ä–µ—Ä–∏–≤–∞—î –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü–∏–∫–ª—É –¥–æ—Å—Ç—Ä–æ–∫–æ–≤–æ.

   –£–º–æ–≤–Ω—ñ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∏ —Ç–∞ —Ü–∏–∫–ª–∏ —à–∏—Ä–æ–∫–æ –∑–∞—Å—Ç–æ—Å–æ–≤—É—é—Ç—å—Å—è –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü—ñ—ó –æ–±—á–∏—Å–ª–µ–Ω—å, –∞–Ω–∞–ª—ñ–∑—É –¥–∞–Ω–∏—Ö, –æ–±—Ä–æ–±–∫–∏ –º–∞—Å–∏–≤—ñ–≤, —Ñ–æ—Ä–º—É–≤–∞–Ω–Ω—è –∑–≤—ñ—Ç—ñ–≤ —Ç–∞ –∫–µ—Ä—É–≤–∞–Ω–Ω—è –ø—Ä–æ—Ü–µ—Å–∞–º–∏.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 1. –û–±—á–∏—Å–ª–µ–Ω–Ω—è –Ω–∞ –æ—Å–Ω–æ–≤—ñ —É–º–æ–≤–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞

```python
x = float(input('–í–≤–µ–¥—ñ—Ç—å —á–∏—Å–ª–æ: '))
if x > 0:
    print('–ß–∏—Å–ª–æ –¥–æ–¥–∞—Ç–Ω–µ')
elif x < 0:
    print('–ß–∏—Å–ª–æ –≤—ñ–¥‚Äô—î–º–Ω–µ')
else:
    print('–¶–µ –Ω—É–ª—å')
```
### üî¢ –ó–∞–≤–¥–∞–Ω–Ω—è:
1. –í–∏–∑–Ω–∞—á–∏—Ç–∏, —á–∏ —î —á–∏—Å–ª–æ –ø–∞—Ä–Ω–∏–º –∞–±–æ –Ω–µ–ø–∞—Ä–Ω–∏–º.

3. –í–∏–∑–Ω–∞—á–∏—Ç–∏, —á–∏ –∫—Ä–∞—Ç–Ω–µ —á–∏—Å–ª–æ 3 –∞–±–æ 5.
4. –í–∏–∑–Ω–∞—á–∏—Ç–∏, —è–∫–µ –∑ –¥–≤–æ—Ö –≤–≤–µ–¥–µ–Ω–∏—Ö —á–∏—Å–µ–ª –±—ñ–ª—å—à–µ.

7. –ü–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏, —á–∏ —î —á–∏—Å–ª–æ –∫–≤–∞–¥—Ä–∞—Ç–æ–º —Ü—ñ–ª–æ–≥–æ —á–∏—Å–ª–∞.
8. –í–∏–∑–Ω–∞—á–∏—Ç–∏, —á–∏ —î –≤–≤–µ–¥–µ–Ω–µ —á–∏—Å–ª–æ –ø—Ä–æ—Å—Ç–∏–º.
9. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—é –≤—ñ–∫—É –ª—é–¥–∏–Ω–∏ (–¥–∏—Ç–∏–Ω–∞, –ø—ñ–¥–ª—ñ—Ç–æ–∫, –¥–æ—Ä–æ—Å–ª–∏–π, –ø–µ–Ω—Å—ñ–æ–Ω–µ—Ä).

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 2. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤–∫–ª–∞–¥–µ–Ω–∏—Ö —É–º–æ–≤
**–ü—Ä–∏–∫–ª–∞–¥:** –ü—Ä–æ–≥—Ä–∞–º–∞, —â–æ –≤–∏–∑–Ω–∞—á–∞—î —Ç–∏–ø —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫–∞ –∑–∞ –¥–æ–≤–∂–∏–Ω–∞–º–∏ —Å—Ç–æ—Ä—ñ–Ω.
```python
a = float(input('a: '))
b = float(input('b: '))
c = float(input('c: '))
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print('–†—ñ–≤–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫')
    elif a == b or a == c or b == c:
        print('–†—ñ–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–∏–π —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫')
    else:
        print('–†—ñ–∑–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω—ñ–π —Ç—Ä–∏–∫—É—Ç–Ω–∏–∫')
else:
    print('–¢—Ä–∏–∫—É—Ç–Ω–∏–∫ –Ω–µ —ñ—Å–Ω—É—î')
```
### üî¢ –ó–∞–≤–¥–∞–Ω–Ω—è:
1. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –ø–æ—Ä—É —Ä–æ–∫—É –∑–∞ –Ω–æ–º–µ—Ä–æ–º –º—ñ—Å—è—Ü—è.
2. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –¥–µ–Ω—å —Ç–∏–∂–Ω—è –∑–∞ —á–∏—Å–ª–æ–º (1‚Äì7).
3. –û–±—á–∏—Å–ª–∏—Ç–∏ –≤–∞—Ä—Ç—ñ—Å—Ç—å –∫–≤–∏—Ç–∫–∞ –∑—ñ –∑–Ω–∏–∂–∫–æ—é –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –≤—ñ–∫—É.
4. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—é —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∏ (–Ω–∏–∑—å–∫–∞, —Å–µ—Ä–µ–¥–Ω—è, –≤–∏—Å–æ–∫–∞).
5. –û–±—á–∏—Å–ª–∏—Ç–∏ –±–æ–Ω—É—Å –¥–æ –∑–∞—Ä–ø–ª–∞—Ç–∏ –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Å—Ç–∞–∂—É.
6. –í–∏–∑–Ω–∞—á–∏—Ç–∏, —á–∏ —î —Å–∏–º–≤–æ–ª –≥–æ–ª–æ—Å–Ω–æ—é —á–∏ –ø—Ä–∏–≥–æ–ª–æ—Å–Ω–æ—é –ª—ñ—Ç–µ—Ä–æ—é.
7. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –ø–æ—Ä—É –¥–æ–±–∏ –∑–∞ –≥–æ–¥–∏–Ω–æ—é.
8. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –≥—Ä—É–ø—É –∫—Ä–æ–≤—ñ –∑–∞ –≤–≤–µ–¥–µ–Ω–∏–º –Ω–æ–º–µ—Ä–æ–º (1‚Äì4).
9. –û–±—á–∏—Å–ª–∏—Ç–∏ —Ç–∞—Ä–∏—Ñ –º–æ–±—ñ–ª—å–Ω–æ–≥–æ –∑–≤‚Äô—è–∑–∫—É –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Ö–≤–∏–ª–∏–Ω.
10. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—é –º–∞—Å–∏ —Ç—ñ–ª–∞ –∑–∞ –Ü–ú–¢.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 3. –¶–∏–∫–ª for
**–ü—Ä–∏–∫–ª–∞–¥:** –í–∏–≤–µ—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–∏ —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ 10.
```python
for i in range(1, 11):
    print(i, i**2)
```
### üî¢ –ó–∞–≤–¥–∞–Ω–Ω—è:
1. –í–∏–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—é –º–Ω–æ–∂–µ–Ω–Ω—è —á–∏—Å–ª–∞ n.
2. –í–∏–≤–µ—Å—Ç–∏ –≤—Å—ñ –ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –¥–æ 50.
3. –û–±—á–∏—Å–ª–∏—Ç–∏ —Å—É–º—É —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ n.
4. –í–∏–≤–µ—Å—Ç–∏ —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª —á–∏—Å–ª–∞ n.
5. –í–∏–≤–µ—Å—Ç–∏ —É—Å—ñ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –¥–æ n.
6. –í–∏–≤–µ—Å—Ç–∏ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤ –ø–µ—Ä—à–∏—Ö n —á–∏—Å–µ–ª.

9. –í–∏–≤–µ—Å—Ç–∏ —á–∏—Å–ª–∞ –§—ñ–±–æ–Ω–∞—á—á—ñ –¥–æ n.
10. –í–∏–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—é –∫—É–±—ñ–≤ —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ n.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 4. –¶–∏–∫–ª while
**–ü—Ä–∏–∫–ª–∞–¥:** –û–±—á–∏—Å–ª–∏—Ç–∏ —Å—É–º—É —á–∏—Å–µ–ª, –ø–æ–∫–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –Ω–µ –≤–≤–µ–¥–µ 0.
```python
s = 0
while True:
    x = float(input('–í–≤–µ–¥—ñ—Ç—å —á–∏—Å–ª–æ (0 ‚Äî –¥–ª—è –≤–∏—Ö–æ–¥—É): '))
    if x == 0:
        break
    s += x
print('–°—É–º–∞:', s)
```
### üî¢ –ó–∞–≤–¥–∞–Ω–Ω—è:
1. –ó–Ω–∞–π—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–≤–µ–¥–µ–Ω–∏—Ö –ø–æ–∑–∏—Ç–∏–≤–Ω–∏—Ö —á–∏—Å–µ–ª.
2. –ó–Ω–∞–π—Ç–∏ –Ω–∞–π–±—ñ–ª—å—à–µ —á–∏—Å–ª–æ —Å–µ—Ä–µ–¥ –≤–≤–µ–¥–µ–Ω–∏—Ö.
3. –ó–Ω–∞–π—Ç–∏ —Å–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ –≤–≤–µ–¥–µ–Ω–∏—Ö —á–∏—Å–µ–ª.
4. –ü—ñ–¥—Ä–∞—Ö—É–≤–∞—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤—ñ–¥‚Äô—î–º–Ω–∏—Ö —á–∏—Å–µ–ª.
5. –í–∏–∑–Ω–∞—á–∏—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–∞—Ä–Ω–∏—Ö —ñ –Ω–µ–ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª.
6. –û–±—á–∏—Å–ª–∏—Ç–∏ –¥–æ–±—É—Ç–æ–∫ —É—Å—ñ—Ö –≤–≤–µ–¥–µ–Ω–∏—Ö —á–∏—Å–µ–ª.
7. –ó–Ω–∞–π—Ç–∏ –º—ñ–Ω—ñ–º–∞–ª—å–Ω–µ —á–∏—Å–ª–æ —Å–µ—Ä–µ–¥ –≤–≤–µ–¥–µ–Ω–∏—Ö.
8. –ü—ñ–¥—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å—É–º—É –ª–∏—à–µ –¥–æ–¥–∞—Ç–Ω–∏—Ö —á–∏—Å–µ–ª.
9. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏, —Å–∫—ñ–ª—å–∫–∏ —Ä–∞–∑—ñ–≤ –≤–≤–µ–¥–µ–Ω–æ –æ–¥–Ω–µ –π —Ç–µ —Å–∞–º–µ —á–∏—Å–ª–æ.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 5. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è break —Ç–∞ continue
**–ü—Ä–∏–∫–ª–∞–¥:** –ü—Ä–æ–ø—É—Å—Ç–∏—Ç–∏ —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ñ 3, —É —Ü–∏–∫–ª—ñ for.
```python
for i in range(1, 11):
    if i % 3 == 0:
        continue
    print(i)
```
### üî¢ –ó–∞–≤–¥–∞–Ω–Ω—è:
1. –ó–∞–≤–µ—Ä—à–∏—Ç–∏ —Ü–∏–∫–ª –ø—Ä–∏ –≤–≤–µ–¥–µ–Ω–Ω—ñ –≤—ñ–¥‚Äô—î–º–Ω–æ–≥–æ —á–∏—Å–ª–∞.

4. –ü—Ä–∏–ø–∏–Ω–∏—Ç–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –ø—Ä–∏ –≤–≤–µ–¥–µ–Ω–Ω—ñ –Ω—É–ª—è.
5. –ü—Ä–æ–ø—É—Å–∫–∞—Ç–∏ —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ñ 7.
6. –ü—Ä–∏–ø–∏–Ω—è—Ç–∏ —Ü–∏–∫–ª, –∫–æ–ª–∏ —Å—É–º–∞ –ø–µ—Ä–µ–≤–∏—â–∏—Ç—å 100.
7. –ü—Ä–æ–ø—É—Å–∫–∞—Ç–∏ —á–∏—Å–ª–∞ –º–µ–Ω—à—ñ –∑–∞ 10.
8. –ó–∞–≤–µ—Ä—à–∏—Ç–∏ —Ü–∏–∫–ª –ø—Ä–∏ –¥–æ—Å—è–≥–Ω–µ–Ω–Ω—ñ –∑–∞–¥–∞–Ω–æ–≥–æ –ø–æ—Ä–æ–≥—É.
9. –ü—Ä–æ–ø—É—Å—Ç–∏—Ç–∏ –≤—Å—ñ –≤—ñ–¥‚Äô—î–º–Ω—ñ —á–∏—Å–ª–∞.
10. –í–∏–≤–µ—Å—Ç–∏ —Ç—ñ–ª—å–∫–∏ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑ –¥—ñ–∞–ø–∞–∑–æ–Ω—É.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 6. –í–∫–ª–∞–¥–µ–Ω—ñ —Ü–∏–∫–ª–∏

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(f'{i*j:3}', end=' ')
    print()
```

**–ü—Ä–∏–∫–ª–∞–¥:** –í–∏–≤–µ—Å—Ç–∏ –ª–∏—à–µ —Ç—ñ —á–∏—Å–ª–∞ –≤—ñ–¥ 1 –¥–æ 50, —è–∫—ñ –∫—Ä–∞—Ç–Ω—ñ 4 —ñ –Ω–µ –∫—Ä–∞—Ç–Ω—ñ 6.

## üìù –ó–∞–≤–¥–∞–Ω–Ω—è 8. –ü—Ä–∞–∫—Ç–∏—á–Ω–µ –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è —Ü–∏–∫–ª—ñ–≤ —Ç–∞ —É–º–æ–≤
**–ü—Ä–∏–∫–ª–∞–¥:** –û–±—á–∏—Å–ª–∏—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª —Å—Ç—É–¥–µ–Ω—Ç–∞ –∑–∞ –≤–≤–µ–¥–µ–Ω–∏–º–∏ –æ—Ü—ñ–Ω–∫–∞–º–∏, –ø–æ–∫–∏ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –Ω–µ –≤–≤–µ–¥–µ -1.
```python
total = 0
count = 0
while True:
    grade = float(input('–û—Ü—ñ–Ω–∫–∞ (-1 ‚Äî –∑–∞–≤–µ—Ä—à–∏—Ç–∏): '))
    if grade == -1:
        break
    total += grade
    count += 1
if count > 0:
    print('–°–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª:', total / count)
else:
    print('–û—Ü—ñ–Ω–æ–∫ –Ω–µ –≤–≤–µ–¥–µ–Ω–æ')
```
